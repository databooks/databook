```python
import tushare as ts
import pandas  as pd
import numpy  as np
import matplotlib.pyplot as plt
from   pylab  import  *
```

```python
df = ts.get_hist_data('300036',ktype='w')
```

```python
df
```

```python
def splot(dfx):
    fig=plt.figure(figsize=(48,4), dpi=80)
    p1=fig.add_subplot(1,1,1)

    p1.plot(dfx.index,dfx['open'])
    p1.plot(dfx.index,dfx['close'])
    p1.plot(dfx.index,dfx['high'])
    p1.plot(dfx.index,dfx['low'])
    plt.grid(True)
    plt.gcf().autofmt_xdate()
    plt.show()
```

```python
def splotbar(dfx):
    fig=plt.figure(figsize=(48,4), dpi=80)
    p1=fig.add_subplot(1,1,1)

    p1.plot(dfx.index,dfx['p_change'], color='red')
    p1.bar(dfx.index,dfx['price_change'])
    plt.grid(True)
    plt.gcf().autofmt_xdate()
    plt.show()
```

```python
df2=df["2019-06-10":"2016-01-01"].sort_index(ascending=True)
```

```python
splot(df2)
```

```python
splotbar(df2)
```

# For Days.

```python
df2018 = ts.get_hist_data('300036',start="2019-05-10",end="2019-06-10")
```

### Price open\high\low\close.

```python
df2018p = pd.DataFrame(df2018,columns=("open","high","low","close")).sort_index(ascending=True)
df2018p.plot(kind='bar',figsize=(32,4),grid=True)
```

### Price variable.

```python
df2018v = pd.DataFrame(df2018,columns=("price_change","p_change")).sort_index(ascending=True)
df2018v.plot(kind='line',figsize=(32,4),grid=True)
df2018v.plot(kind='bar',figsize=(32,4),grid=True)
```

### Price variable cumsum.

```python
df2018c=df2018v.cumsum()
#df2018c = pd.DataFrame(df2018c,columns=("price_change","p_change")).sort_index(ascending=True)
df2018c.plot(kind='line',figsize=(32,4),grid=True)
df2018c.plot(kind='bar',figsize=(32,4),grid=True)
```

```python
def plt_vari_price(dfx):
    fig=plt.figure(figsize=(48,6), dpi=80)
    p1=fig.add_subplot(1,1,1)

    p1.plot(dfx.index,dfx['p_change'])
    p1.bar(dfx.index,dfx['price_change'],color="red")
    
    plt.grid(True)
    plt.gcf().autofmt_xdate()
    plt.show()
```

```python
plt_vari_price(df2018v)
plt_vari_price(df2018c)
```

```python
def plt_index_change(xindex):
    fig=plt.figure(figsize=(48,6), dpi=80)
    p2=fig.add_subplot(1,1,1)

    p2.plot(df2018c.index,df2018c[xindex],color="red")
    p2.bar(df2018v.index,df2018v[xindex])

    plt.grid(True)
    plt.gcf().autofmt_xdate()
    plt.show()   
```

```python
plt_index_change("p_change")
plt_index_change("price_change")    
```
